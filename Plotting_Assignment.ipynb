```python
# YOUR CODE HERE
plt.figure(figsize=(10, 6))
bin_sizes = [10, 20, 30, 40]
for bin_size in bin_sizes:
    sns.histplot(data=data, x='co2', bins=bin_size, kde=True, label=f'Bin Size: {bin_size}')
plt.title('Histogram of Carbon Dioxide Emissions (co2)')
plt.xlabel('Carbon Dioxide Emissions')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.savefig('co2_graph.png')
plt.show()
```

```python
# YOUR CODE HERE
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='cylinders', bins=10, kde=True, color='salmon')
plt.title('Histogram of Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('cylinders_histogram.png')
plt.show()
```

```python
# YOUR CODE HERE

mean_co2_by_year = data.groupby('year')['co2'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=mean_co2_by_year, x='year', y='co2', marker='o', color='skyblue')
plt.title('Change in Carbon Dioxide Emissions (co2) Over the Years')
plt.xlabel('Year')
plt.ylabel('Mean Carbon Dioxide Emissions')
plt.grid(True)
plt.savefig('co2_year_lineplot.png')
plt.show()
```
